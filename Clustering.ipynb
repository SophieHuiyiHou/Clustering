# Data preprocessing and getting to know the data

### Because the values reported are standardised values of original values, there is no need to apply scaling before performing clustering. Here we directly go for clustering.

```python
# take 5 variables and drop the rest
df3 = df_mar[['residual_transmission', 'regulatable_transmission', 'recovery_rate','detection_rate','death_rate']]

print(df3.info())

# convert df_mar to matrix
X_mar = df3.to_numpy()
```

```python
# take 5 variables and drop the rest
df4 = df_apr[['residual_transmission', 'regulatable_transmission', 'recovery_rate','detection_rate','death_rate']]

print(df4.info())

# convert df_apr to matrix
X_apr = df4.to_numpy()
```

# Using elbow method to determine K

```python
from sklearn.cluster import KMeans

# random state, we will use 42
rs = 42
```

```python
# list to save the clusters and cost
clusters = []
inertia_vals = []

# this whole process should take a while, K from 2 to 15, with intervel = 2.
for k in range(2, 15, 2):
    # train clustering with the specified K
    model = KMeans(n_clusters=k, random_state=rs)
    model.fit(X_mar)
    
    # append model to cluster list
    clusters.append(model)
    inertia_vals.append(model.inertia_)
```

```python
# plot the inertia vs K values
plt.plot(range(2,15,2), inertia_vals, marker='*')
plt.show()
```

```python
# list to save the clusters and cost
clusters = []
inertia_vals = []

# this whole process should take a while, K from 2 to 15, with intervel = 2.
for k in range(2, 15, 2):
    # train clustering with the specified K
    model = KMeans(n_clusters=k, random_state=rs)
    model.fit(X_apr)
    
    # append model to cluster list
    clusters.append(model)
    inertia_vals.append(model.inertia_)
```

```python
# plot the inertia vs K values
plt.plot(range(2,15,2), inertia_vals, marker='*')
plt.show()
```

## For March, optimal 𝐾 could be either 4 or 8.
## For April, optimal 𝐾 could be either 4 or 6.

# Calculating silhouette score to determine the optimal K

```python
from sklearn.metrics import silhouette_score

print(clusters[1])
print("Silhouette score for k=4", silhouette_score(X_mar, clusters[1].predict(X_mar)))

print(clusters[3])
print("Silhouette score for k=8", silhouette_score(X_mar, clusters[3].predict(X_mar)))
```

```python
print(clusters[1])
print("Silhouette score for k=4", silhouette_score(X_apr, clusters[1].predict(X_apr)))

print(clusters[2])
print("Silhouette score for k=6", silhouette_score(X_apr, clusters[2].predict(X_apr)))
```

## For March: Silhouette score for k=4 is 0.42 and Silhouette score for k=8 is 0.28. Therefore k=4 is the opitmal K.
## For April: Silhouette score for k=4 is 0.36 and Silhouette score for k=6 is 0.22. Therefore k=4 is the opitmal K.

# Final cluster solution

```python
# visualisation of K=4 clustering solution
model_mar = KMeans(n_clusters=4, random_state=rs)
model_mar.fit(X_mar)

# sum of intra-cluster distances
print("Sum of intra-cluster distance:", model_mar.inertia_)

print("Centroid locations:")
for centroid in model_mar.cluster_centers_:
    print(centroid)

y = model_mar.predict(X_mar)
df3['Cluster_ID'] = y

# how many in each
print("Cluster membership")
print(df3['Cluster_ID'].value_counts())

# pairplot
# added alpha value to assist with overlapping points
cluster_g = sns.pairplot(df3, hue='Cluster_ID', diag_kind='hist', palette = 'rainbow') # or palette = 'rainbow' or 'dark'
plt.show()
```

```python
# visualisation of K=4 clustering solution
model_apr = KMeans(n_clusters=4, random_state=rs)
model_apr.fit(X_apr)

# sum of intra-cluster distances
print("Sum of intra-cluster distance:", model_apr.inertia_)

print("Centroid locations:")
for centroid in model_apr.cluster_centers_:
    print(centroid)

y = model_apr.predict(X_apr)
df4['Cluster_ID'] = y

# how many in each
print("Cluster membership")
print(df4['Cluster_ID'].value_counts())

# pairplot
# added alpha value to assist with overlapping points
cluster_g = sns.pairplot(df4, hue='Cluster_ID', diag_kind='hist', palette = 'rainbow')
plt.show()
```
